```python
import cv2
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
img = cv2.resize(img,(600,600),interpolation=cv2.INTER_CUBIC)
```

```python
# blue-ish output because matplotlib uses RGB instead of BGR in opencv
plt.imshow(img)
```

```python
# Get list of images, #change the path to your target images path
images = glob.glob("images/*.jpg")
len(images)
```

```python
frames = []
for idx,i in enumerate(range(0,len(images))):
    img = cv2.imread(images[i])
    img = cv2.resize(img,(600,600),interpolation=cv2.INTER_CUBIC)
    frames.append(img)
    print(idx)
print("Done!")
```

```python
#plot to see if images are stored inside frames list
plt.imshow(frames[4])
```

```python
# set some parameters for video writer
fps = 1
size = (600,600)
output_filename = "output.avi"
out = cv2.VideoWriter(output_filename,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
```

```python
for i in range(len(frames)):
    # writing to a image array
    out.write(frames[i])
out.release()
```
