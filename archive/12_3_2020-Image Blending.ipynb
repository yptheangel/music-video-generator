```python
import cv2
import matplotlib.pyplot as plt
```

```python
ori=cv2.imread("data/ori.jpg")
styled=cv2.imread("data/styled.jpg")
```

```python
plt.imshow(cv2.cvtColor(ori,cv2.COLOR_BGR2RGB))
```

```python
ori.shape
```

```python
styled=cv2.resize(styled,(ori.shape[1],ori.shape[0]))
```

```python
styled.shape
```

```python
plt.imshow(cv2.cvtColor(styled,cv2.COLOR_BGR2RGB))
```

```python
alpha = 0.6
beta = (1.0 - alpha)
output = cv2.addWeighted(ori, alpha, styled, beta, 0.0)
```

```python
plt.imshow(cv2.cvtColor(output,cv2.COLOR_BGR2RGB))
```
